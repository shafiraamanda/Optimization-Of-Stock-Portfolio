# Saham Konsisten LQ45 Tahun 2025

```python
import pandas as pd

# 1. Baca file Excel (data 2023–2025)
file_path = r"D:\Semester 1-7\SKRIPSI\5. Data\Data 2023-2025\Daftar Saham Konsisten LQ45 Periode 2023-2025.xlsx"
df = pd.read_excel(file_path)

# 2. Ambil semua kolom evaluasi (2023–2025)
eval_cols = df.columns.tolist()   # otomatis ambil semua kolom

# 3. Cari saham yang muncul di SEMUA periode (konsisten 2023–2025)
consistent_stocks = set(df[eval_cols[0]].dropna())

for col in eval_cols[1:]:
    consistent_stocks = consistent_stocks.intersection(set(df[col].dropna()))

# 4. Urutkan hasil
tickers = sorted(consistent_stocks)

# 5. Tampilkan hasil
print("Saham yang konsisten masuk LQ45 selama periode 2023–2025:")
print(tickers)

print(f"\nTotal saham konsisten: {len(tickers)}")
```

# Scrape Saham (Data Historis)

```python
import yfinance as yf
import pandas as pd
import os


tickers = [t + ".JK" for t in tickers]  # tambahkan .JK sekali saja

# Lokasi file output
output_path = r"D:\Semester 1-7\SKRIPSI\5. Data\Data 2023-2025\stocks_data_2023-2025.xlsx"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

with pd.ExcelWriter(output_path) as writer:
    for ticker in tickers:
        try:
            print(f"Downloading {ticker} ...")
            df = yf.download(
                ticker,
                start="2023-01-01",
                end="2025-12-31",
                interval="1d",
                auto_adjust=False
            )

            if df.empty:
                print(f"Warning: {ticker} returned empty dataframe, skipping.")
                continue

            df = df.reset_index()

            # --- Flatten MultiIndex columns jika ada ---
            df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]

            # --- Standardisasi nama kolom ---
            rename_dict = {
                'Date': 'Date',
                'Adj Close': 'Adj Close',
                'Adj Close_': 'Adj Close',
                'Close': 'Close',
                'Close_': 'Close',
                'High': 'High',
                'High_': 'High',
                'Low': 'Low',
                'Low_': 'Low',
                'Open': 'Open',
                'Open_': 'Open',
                'Volume': 'Volume',
                'Volume_': 'Volume'
            }

            # Ambil bagian sebelum _ jika ada, dan ganti sesuai standar
            df.rename(columns=lambda x: rename_dict.get(x.split('_')[0], x.split('_')[0]), inplace=True)

            # Tambahkan kolom Ticker
            df["Ticker"] = ticker.replace(".JK", "")

            # Sheet name maksimal 31 karakter
            df.to_excel(
                writer,
                sheet_name=ticker.replace(".JK", "")[:31],
                index=False,
                float_format="%.2f"
            )

        except Exception as e:
            print(f"Failed to download {ticker}: {e}")

print("Download selesai! File tersimpan di:", output_path)
```
